```python
#Calcul de la circulation du vortex
Up = np.linspace(2,7,100) #Vitesse du piston en m.s-1
tp = None #Temps de course du piston
Ln = 250e-3 #Course du piston en m
Dn = np.linspace(50,100,100)*1e-3 #Diamètre du piston en m
mu = 15.6e-6 #viscosité cinématique de l'air pour 25°C
#print(Ln/Dn)
plt.figure()

for i in [0,25,50,75,99]:
    tp = Ln/Up
    GAMA = (Up**2)*(tp/2)
    D = Dn[i]*1.18*(Ln/Dn[i])**0.25
    U = (GAMA/(2*np.pi*D))*(np.log((4*D)/((4*mu*tp)**0.5))-0.558)
    R = D/2
    xi = R*(np.exp(GAMA/(4*np.pi*R))/np.exp(U))

    #print('La circulation du vortex Gama = {}, le diamètre du vortex D = {}, La vitesse du vortex U = {} et épaisseur du vortex xi = {}'.format(GAMA,D,U,xi))

    #plt.figure()
    #plt.plot(Up,D)
    plt.plot(Up,U,label="Dn = "+str(Dn[i]))
plt.legend()
plt.figure()
for i in [0,25,50,75,99]:
    tp = Ln/Up
    GAMA = (Up**2)*(tp/2)
    D = Dn[i]*1.18*(Ln/Dn[i])**0.25
    U = (GAMA/(2*np.pi*D))*(np.log((4*D)/((4*mu*tp)**0.5))-0.558)
    R = D/2
    xi = R*(np.exp(GAMA/(4*np.pi*R))/np.exp(U))

    #print('La circulation du vortex Gama = {}, le diamètre du vortex D = {}, La vitesse du vortex U = {} et épaisseur du vortex xi = {}'.format(GAMA,D,U,xi))

    #plt.figure()
    #plt.plot(Up,D)
    plt.plot(Up,xi,label="Dnl = "+str(Dn[i]))
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

n = 100
for i in [0,25,50,75,99]:
    theta = np.linspace(0, 2.*np.pi, n)
    phi = np.linspace(0, 2.*np.pi, n)
    theta, phi = np.meshgrid(theta, phi)
    D = Dn[i]*1.18*(Ln/Dn[i])**0.25
    c, a = D, xi[i]/2
    x = (c + a*np.cos(theta)) * np.cos(phi)
    y = (c + a*np.cos(theta)) * np.sin(phi)
    z = a * np.sin(theta)

    fig = plt.figure(figsize=((20,10)))
    ax1 = fig.add_subplot(121, projection='3d')
    ax1.set_xlim(-0.1,0.1)
    ax1.set_ylim(-0.1,0.1)
    ax1.set_zlim(-0.1,0.1)
    ax1.plot_surface(x, y, z, rstride=5, cstride=5, color='r')#, edgecolors='w')
    ax1.view_init(36, 26)
    ax2 = fig.add_subplot(122, projection='3d')
    ax2.set_xlim(-0.1,0.1)
    ax2.set_ylim(-0.1,0.1)
    ax2.set_zlim(-0.1,0.1)
    ax2.plot_surface(x, y, z, rstride=5, cstride=5, color='r')#, edgecolors='w')
    ax2.view_init(0, 0)
    ax2.set_xticks([])
    fig.savefig("cannon{}.png".format(i))
    plt.title("vitesse Up = {}".format(Up[i]))
    fig.show()
```
